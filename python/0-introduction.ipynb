## 対話モードで遊ぶ

`python3`というコマンドをターミナルに入力すると、Pythonインタプリタが**対話モード**で起動します。対話モードを使うと、プログラムを入力して、`Enter`キーを押してすぐに実行することが出来ます。

対話モードを終了するには`exit()`と入力して`Enter`キーを押すか、`Ctrl`キーと`d`キーを同時に押してください。(`Ctrl + d`)

例

```
[shuto@laptop ~]$ python3
Python 3.10.6 (main, Mar 10 2023, 10:55:28) [GCC 11.3.0] on linux
Type "help", "copyright", "credits" or "license" for more information.
>>> "Hello world"
Hello world
>>> exit()
```

### 電卓

Pythonインタプリタを簡単な電卓として使ってみましょう。 \
計算したい式(プログラム)を入力して、`Enter`キーを押すとプログラムを実行した結果が表示されます。間違ったプログラムを実行すると**エラー**の内容が表示されます。

Pythonでは`+`, `-`, `*`, `/`などの**演算子**が使えます。

例 (`#`の部分は入力しなくてもいいです。)

```
>>> 1 + 1 # 加算
2
>>> 123 - 45 # 減算
78
>>> 2 * 9 # 乗算
18
>>> 7 / 2 # 除算
3.5
>>> 7 // 2 # 除算 (小数点以下を切り捨て)
3
>>> 11 % 3 # 余剰 (余り)
2
>>> 2 ** 8 # 2の8乗
256
>>> (1 + 2) * (3 + 4) # 丸括弧
21
>>> 1.42 + 5.49 # 小数
6.91
>>> 4 / 0
Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
ZeroDivisionError: division by zero
>>> 3 +
  File "<stdin>", line 1
    3 +
       ^
SyntaxError: invalid syntax
```

### 文字列

文字列は`"おはよう"`や`"こんにちは"`のように引用符を使って書きます。文字列は`+`を使って、2つの文字列を結合することも出来ます。

例
```
>>> "こんにちは！"
'こんにちは！'
>>> "マックに行きませんか？"
'マックに行きませんか？'
```

## スクリプトを実行する

実際にPythonのプログラムを作成するときは予めファイルに書いておいて、そのファイルをPythonインタプリタで実行することが多いです。Pythonプログラムはファイルの末尾に`.py`をつけます。(例 `filename.py`)

ファイルを作ってみましょう。`hello.py`と名付けて以下の内容を書いてください。

```python
print("こんにちは")
```

ファイルを作ったら実行してみましょう。ターミナルで以下のように実行してみてください。

```
python3 hello.py
```

画面には`こんにちは`と表示されているはずです。 \
対話モード以外で画面に文字や数を画面に表示させるためには`print()`関数を使います。

### コメント

`#`以降に書かれた部分は**コメント**と呼ばれ、プログラムとしての意味はもちません。プログラマーがメモをプログラム中に残すために使われます。他にもプログラムの一部分を実行させないときに使ったりします。

```python
# こんにちはと表示する
print("こんにちは！")

# お昼ご飯を吉野家に誘おう
# 南草津駅付近には吉野家がないので行けない…
# print("吉野家に行きませんか？")

# じゃあマックに誘おう
print("マックに行きませんか？")
```
